```python
def split_image(foto, row_sizes, col_sizes):

  pecahan = []

  for i in range(1, len(row_sizes)):
    for j in range(1, len(col_sizes)):
      chunk = foto[sum(row_sizes[1:i]):sum(row_sizes[1:i+1]), sum(col_sizes[1:j]):sum(col_sizes[1:j+1])]
      pecahan.append(chunk)

  return pecahan

row_sizes = [0, 256, 256, 256, 256, 256, 256, 256, 256] 
col_sizes = [0, 1026, 1026]
chunks = split_image(foto, row_sizes, col_sizes)
```

```python
def translation_img(img, x=0, y=0):
    height, width = img.shape
    result = np.zeros((height, width))
    for i in range(height):
        for j in range(width):
            result[i, j] = img[i - y, j - x]
    return result

hasil_translasi = []

hasil_translasi.append(translation_img(chunks[2], 513, 0))
hasil_translasi.append(translation_img(chunks[3], 513, 0))
hasil_translasi.append(translation_img(chunks[4], 513, 0))
hasil_translasi.append(translation_img(chunks[5], 513, 0))
hasil_translasi.append(translation_img(chunks[10], 513, 0))
hasil_translasi.append(translation_img(chunks[11], 513, 0))
hasil_translasi.append(translation_img(chunks[12], 513, 0))
hasil_translasi.append(translation_img(chunks[13], 513, 0))
```

```python
def rotation_img(image, degr=0):
    height, width = image.shape
    hasil = np.zeros((height,width))
    radian =  np.pi * (degr / 180)
    for i in range(height):
        for j in range(width):
            new_x = int((i-height/2) * np.cos(radian) - (j-width/2) * np.sin(radian) + height/2)
            new_y = int((i-height/2) * np.sin(radian) + (j-width/2) * np.cos(radian) + width/2)
            if (0 <= new_x < height and 0 <= new_y < width):
                hasil[i,j] = image[new_x, new_y]
    return hasil
```

Code diatas merupakan sebuah fungsi rotasi yang dimana fungsi ini akan merotasikan citra sesuai dengan `degr` atau derajat yang dimasukkan kedalam parameternya. Kemudian derajat sudut diubah menjadi derajat radian dengan menggunakan library numpy yaitu `np.pi()` yang diisi dengan `degr / 180`. Selanjutnya terdapat dua kali perulangan, perulangan pertama untuk mengiterasi melalui indeks baris dari `0 hingga height - 1` dan perulangan kedua untuk mengiterasi melalui indeks kolom dari `0 hingga width - 1`.

```python
def sorting(array, index):
    sort = []
    for i in index:
        sort.append(array[i])
    return sort

index = [0, 2, 4, 6, 1, 3, 5, 7]
sort = sorting(hasil, index)

for i in range(len(sort)):
  sort[i] = sort[i][ : , 215:-215]
  new_sorting_result = np.zeros((sort[i].shape[0] + 390, sort[i].shape[1]))
  new_sorting_result[390:, :] = sort[i]
  sort[i] = new_sorting_result
  sort[i] = rotation_img(sort[i], 90)
  sort[i] = sort[i][:, 2:-388]
  
for i in range(len(sort)):
  if sort[i].shape[0] != 200:
     sort[i] = sort[i][:, :-4]
```

Code dibawah merupakan sebuah fungsi menggabungkan pecahan-pecahan citra menjadi citra yang utuh sepenuhnya. Fungsi `merge_img()` menggunakan parameter `(array, row, col)`, yang dimana untuk gabungan gambar nya itu mulai dari index pertama hingga terakhir dari panjang array dengan `row=2` dan `col=4`. Untuk citra yang telah digabungkan ini, akan ditampilkan tanpa perluangan, dengan menggunakan `imshow()` yang diisi dengan variabel `gambar_gabungan`.

```python
def merge_img(array, row, col):
    max_row = max([img.shape[0] for img in array])
    max_col = max([img.shape[1] for img in array])
    new_canvas = np.zeros((row * max_row, col * max_col))
    x = 0
    y = 0
    for i in range(len(array)):
        height = array[i].shape[0]
        width = array[i].shape[1]

        new_canvas[y:y + height, x:x + width] = array[i]

        if (i + 1) % col == 0:
            x = 0
            y += height
        else:
            x += width

    return new_canvas

gambar_gabungan = merge_img(sort, 2, 4)

plt.imshow(gambar_gabungan, cmap='gray')
plt.axis("off")
plt.show()
```

```python
grayValue = [i for i in range(256)]
panjang, lebar = gambar_gabungan.shape

def histogram(image):
    sumGray = np.zeros(256).astype(int)
    
    for i in range(panjang):
        for j in range(lebar):
            value = image[i, j]
            sumGray[int(value)] += 1
    return sumGray
```


Code diatas adalah implementasi fungsi histogram untuk menghitung distribusi intensitas pixel dalam gambar. Menggunakan array `sumGray`, program menghitung jumlah kemunculan tiap nilai intensitas pixel dari `0 hingga 255`. Proses ini dilakukan dengan iterasi melalui setiap pixel dalam gambar dan menambahkan frekuensi kemunculan nilai intensitas pixel ke array sesuai dengan nilai pixel yang ditemukan. Hasilnya adalah array yang merepresentasikan histogram gambar, di mana setiap indeks menyimpan jumlah pixel dengan nilai intensitas tertentu.

Code dibawah merupakan implementasi dari proses equalisasi histogram pada gambar. Pertama, menghitung histogram dari gambar yang diberikan dan kemudian menghitung probabilitas munculnya setiap nilai intensitas pixel. Selanjutnya, menghitung distribusi kumulatif probabilitas pixel, yang merepresentasikan transformasi fungsi intensitas yang akan diterapkan pada gambar. Dengan menggunakan distribusi kumulatif, program melakukan proses equalisasi histogram pada gambar, di mana setiap pixel dalam gambar diberi nilai baru berdasarkan transformasi yang dihasilkan. Hasilnya adalah gambar yang telah diequalisasi histogramnya. Terakhir, kode menampilkan gambar asli dan gambar yang telah diequalisasi histogramnya secara berdampingan menggunakan matplotlib. Hal ini memberikan visualisasi perbandingan antara gambar asli dan gambar yang telah diequalisasi histogramnya.

```python
height, width = gambar_gabungan.shape
sumGray = histogram(gambar_gabungan)
probabilitas = np.zeros(256)

for i in range(256):
    probabilitas[i] = sumGray[i] / (height * width)

sumNew = np.zeros(256)
sumNew[0] = probabilitas[0]

for i in range(1, 256):
    sumNew[i] = sumNew[i - 1] + probabilitas[i]

gambar_gabungan_equalized = gambar_gabungan.copy()
for i in range(height):
    for j in range(width):
        pixel = gambar_gabungan[i, j]
        newPixel = round(sumNew[int(pixel)] * 255)
        gambar_gabungan_equalized[i, j] = newPixel

plt.subplot(1,2,1)
plt.imshow(gambar_gabungan, cmap='gray')
plt.title('Image Real')
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(gambar_gabungan_equalized, cmap='gray')
plt.title('Equalized Merged Image')
plt.axis("off")
plt.show()
```

```python
def img_specification(first_img, second_img):
    histogram1, _ = np.histogram(first_img.flatten(), bins=256, range=(0,256))
    histogram2, _ = np.histogram(second_img.flatten(), bins=256, range=(0,256))

    normalisasi1 = histogram1 / (first_img.shape[0] * first_img.shape[1])
    normalisasi2 = histogram2 / (second_img.shape[0] * second_img.shape[1])

    cdf1 = np.cumsum(normalisasi1)
    cdf2 = np.cumsum(normalisasi2)

    mapping = np.zeros(256)
    for i in range(256):
        j = 0
        while cdf2[j] < cdf1[i] and j < 255:
            j += 1
        mapping[i] = j

    spesification_img = np.zeros_like(first_img)
    for i in range(first_img.shape[0]):
        for j in range(first_img.shape[1]):
            spesification_img[i, j] = mapping[int(first_img[i, j])]

    return spesification_img
```


Code diatas merupakan Fungsi `img_specification` yang menerapkan metode spesifikasi histogram untuk mengubah distribusi intensitas pixel dari gambar pertama `first_img` agar sesuai dengan distribusi intensitas pixel dari gambar kedua `second_img`. Pertama, fungsi menghitung histogram normalisasi dari kedua gambar dan menghitung distribusi kumulatifnya `CDF`. Kemudian, untuk setiap nilai intensitas pixel dalam gambar pertama, fungsi mencari nilai yang sesuai dalam CDF gambar kedua dan memetakan nilai intensitas pixel tersebut. Dengan menggunakan pemetaan ini, fungsi menghasilkan gambar yang telah disesuaikan histogramnya `spesification_img`. Hal ini memungkinkan penyesuaian distribusi intensitas pixel antara dua gambar, sering digunakan dalam pemrosesan citra untuk memperbaiki kontras atau menyesuaikan pencahayaan antara gambar-gambar yang berbeda.

Code dibawah akan menampilkan empat gambar dalam satu bingkai menggunakan library matplotlib. Gambar pertama adalah gambar awal, yang kedua adalah gambar asli, yang ketiga adalah hasil dari spesifikasi histogram yang diterapkan pada gambar asli agar sesuai dengan gambar target `kurumi`, dan yang keempat adalah gambar target. Setiap gambar disajikan dalam `subplot()` yang sesuai, dengan judul yang sesuai dengan jenis gambarnya, dan sumbunya dihilangkan agar tampilan lebih bersih.
```python
def fracSpec(images, row=1, col=1):
    img_list = []
    batasan = 128

    fracRow = [0] + [images.shape[0] // row] * row
    fracCol = [0] + [images.shape[1] // col] * col

    fracImg = split_image(images, fracRow, fracCol)

    lenRow = len(fracRow) - 1
    lenCol = len(fracCol) - 1

    for i in range(len(fracImg)):
        rata_rata_pixel = np.mean(fracImg[i])
        if rata_rata_pixel <= batasan:
            img_list.append(img_specification(fracImg[i], gray_img2))
        else:
            img_list.append(img_specification(fracImg[i], gray_img3))
    
    hasil_akhir = merge_img(img_list, lenRow, lenCol)
    return hasil_akhir

fracSpec_img = []
fracSpec_img.append(fracSpec(gambar_gabungan_equalized, 1, 1))
fracSpec_img.append(fracSpec(gambar_gabungan_equalized, 1, 2))
fracSpec_img.append(fracSpec(gambar_gabungan_equalized, 2, 2))
fracSpec_img.append(fracSpec(gambar_gabungan_equalized, 2, 16))
fracSpec_img.append(fracSpec(gambar_gabungan_equalized, 16, 16))
```

Setelah itu, kode menggunakan fungsi ini untuk menghasilkan serangkaian gambar dengan variasi pengaturan jumlah potongan gambar.
