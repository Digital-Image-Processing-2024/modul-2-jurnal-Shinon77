```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
import pandas as pd
```

```python
foto = plt.imread("Res/senheScramble.jpg")
plt.imshow(foto, cmap='gray')
tinggi,lebar = foto.shape
tinggi2 = int(tinggi / 2)
lebar2 = int(lebar / 2)
```

```python
rotation_corner = 270

sin_theta = np.sin(np.radians(rotation_corner))
cos_theta = np.cos(np.radians(rotation_corner))

rotation_image = np.zeros_like(foto)
for i in range(tinggi):
    for j in range(lebar):
        newX = int(round(cos_theta * (j - foto.shape[1] // 2) - sin_theta * (i - foto.shape[0] // 2) + foto.shape[1] // 2))
        newY = int(round(sin_theta * (j - foto.shape[1] // 2) + cos_theta * (i - foto.shape[0] // 2) + foto.shape[0] // 2))

        if 0 <= newX < foto.shape[1] and 0 <= newY < foto.shape[0]:
            rotation_image[i, j] = foto[newY, newX]

plt.title('Rotation Image')
plt.imshow(rotation_image, cmap='gray')
```
